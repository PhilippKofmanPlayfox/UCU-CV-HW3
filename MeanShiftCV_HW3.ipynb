```
# !unzip  s-originals.zip
import cv2 as cv
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
import numpy as np
```

```
slabels = []
for i in range(1, 5):
  x = pd.read_csv(f"s-originals/s{i}-cb.txt", sep='\s+', header=None)
  slabels.append(x)
  
sdata = []
for i in range(1, 5):
  x = pd.read_excel(f"s-originals/s{i}-groundtruth-plot.xls", sep='\s+', header=None)
  sdata.append(x)
```

# Dataset S

```
plt.figure(figsize=(11, 6))
for i in range(4):
    clustering = MeanShift(bandwidth=100000).fit(sdata[i].iloc[:, [1,2]].values)
    centrs = clustering.cluster_centers_
    
    plt.figure(figsize=(11, 6))
    plt.subplot(1, 2, 1)
    plt.scatter(sdata[i].iloc[:, 1], sdata[i].iloc[:, 2], color='green', alpha=0.7)
    plt.scatter(slabels[i].iloc[:, 0], slabels[i].iloc[:, 1], color='red', alpha=0.9)
    
    plt.title('REAL')
    
    plt.subplot(1, 2, 2)
    plt.scatter(sdata[0].iloc[:, 1], sdata[0].iloc[:, 2], color='green', alpha=0.7)
    plt.scatter(centrs[:, 0], centrs[:, 1], color='red', alpha=0.9)
    plt.title('Mean-Shift')
    plt.show()
```
